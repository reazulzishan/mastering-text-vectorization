One Hot Encoding


```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample categorical data
data = np.array([['Red'], ['Green'], ['Blue'], ['Red'], ['Blue']])

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False) # sparse_output=False returns an array

# Fit and transform the data
encoded_data = encoder.fit_transform(data)
```

N-Grams

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                           "dswithbappy watch dswithbappy",
                           "people write comment",
                           "dswithbappy write comment"], "output":[1,1,0,0]})

df
```

```python
#BI-gram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df["text"])
```

```python
bow.toarray()
```

```python
#vocavulary
print(cv.vocabulary_)
```

```python
#TI-gram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df["text"])
```

```python
bow.toarray()
```

```python
#vocavulary
print(cv.vocabulary_)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import os
story = []

for filename in os.listdir('DATA'):
    if filename.endswith('.txt'):
        with open(os.path.join('DATA', filename), 'r', encoding='utf-8', errors='ignore') as f:
            corpus = f.read()
            raw_sent = sent_tokenize(corpus)
            for sent in raw_sent:
                story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python

model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```
